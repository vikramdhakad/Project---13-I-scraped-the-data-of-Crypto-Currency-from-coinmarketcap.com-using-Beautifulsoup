```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from lxml import etree
```

```python
url = 'https://coinmarketcap.com/all/views/all/'
request = requests.get(url)
```

```python
soup = BeautifulSoup(request.content, "html.parser")
```

```python
dom = etree.HTML(str(soup)) 
```

```python
name = []
symbol = []
market_ca = []
prices = []

for names in soup.find_all("a", class_="cmc-table__column-name--name cmc-link"):
    name.append(names.text)

for sym in soup.find_all("a", class_="cmc-table__column-name--symbol cmc-link"):
    symbol.append(sym.text)

for market_cap in soup.find_all("span", class_="sc-7bc56c81-0 dCzASk"):
    market_ca.append(market_cap.text)

for price in range(0,20):
    prices.append(dom.xpath("//a[@class='cmc-link']/span")[price].text)
```

```python
data = pd.DataFrame()
data['Crypto Currency Name'] = name
data['Currency Symbol'] = symbol
data['Market Capital'] = market_ca
data['Rate'] = prices

data
```

```python
data.to_csv('Crypto Currenies Data.csv')
```
